```python
# Load and normalize per site
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("era5_features_sites.csv", parse_dates=["valid_time"])
# features = ["t2m", "tcc", "ssrd", "q", "t", "cc", "u", "v"]
features = ["ssrd", "q", "t", "cc", "u", "v"]
site_groups = dict(tuple(df.groupby("site")))

# Normalize each site independently
scalers = {}
for site, data in site_groups.items():
    scaler = MinMaxScaler()
    site_groups[site][features] = scaler.fit_transform(data[features])
    scalers[site] = scaler
```

```python
import numpy as np

def create_sequences(data, input_len=16, output_len=1):
    X, y = [], []
    for i in range(len(data) - input_len - output_len + 1):
        X.append(data[i:i+input_len])
        y.append(data[i+input_len:i+input_len+output_len])
    return np.array(X), np.array(y)

X_all, y_all = [], []
for site, data in site_groups.items():
    X, y = create_sequences(data[features].values, input_len=16)
    X_all.append(X)
    y_all.append(y)

X_all = np.concatenate(X_all)
y_all = np.concatenate(y_all)
```
