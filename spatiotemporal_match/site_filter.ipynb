### single-level only

```python
import xarray as xr
import pandas as pd
import numpy as np
import os

# === path to your single-level (accumulated) ERA5 file with SSRD ===
accum_file = "/mnt/e/CERA/CausalST-forecast/6ee3307ab5f4ad1a21b5337a13effb9e.nc"

# === study sites ===
SOLAR_FIELDS = [
    {"name": "Bhadla",    "lat": 27.5833,  "lon":  71.4333},
    {"name": "Tengger",   "lat": 37.3333,  "lon": 103.8558},
    {"name": "Benban",    "lat": 24.5561,  "lon":  32.9016},
    {"name": "Cestas",    "lat": 44.7255,  "lon":  -0.8157},
    {"name": "Balboa",    "lat": 38.4533,  "lon":  -6.2260},   # Núñez de Balboa
    {"name": "Topaz",     "lat": 35.2426,  "lon": -120.0096},
    {"name": "DesertSun", "lat": 33.8214,  "lon": -115.3939},
    {"name": "Pirapora",  "lat": -17.0891, "lon": -44.9878},
    {"name": "Nyngan",    "lat": -31.5575, "lon": 147.2031},
]

# --- open and normalize time coordinate ---
def open_ssrd_dataset(path):
    ds = xr.open_dataset(path)  # decode_times=True by default
    # unify the time coordinate name to 'time'
    if 'valid_time' in ds.coords and 'time' not in ds.coords:
        ds = ds.rename({'valid_time': 'time'})
    return ds

ds = open_ssrd_dataset(accum_file)
assert 'ssrd' in ds.data_vars, "The dataset must contain the 'ssrd' variable."

# --- helper: ERA5 longitudes can be 0..360 or -180..180 ---
def normalize_lon_for_ds(lon, ds):
    if 'longitude' not in ds.coords:
        return lon
    lon_max = float(ds['longitude'].max())
    return lon % 360 if lon_max > 180 else lon

# --- extract only SSRD at the nearest grid point for each site ---
def extract_ssrd_site(ds, site):
    lat = site['lat']
    lon = normalize_lon_for_ds(site['lon'], ds)
    sub = ds.sel(latitude=lat, longitude=lon, method='nearest')
    df = sub[['ssrd']].to_dataframe().reset_index()
    df['site'] = site['name']
    # keep selected gridpoint coords (nice for sanity checks)
    df['latitude']  = float(sub['latitude'])
    df['longitude'] = float(sub['longitude'])
    return df[['time', 'site', 'latitude', 'longitude', 'ssrd']]

# collect all sites
all_sites_df = pd.concat([extract_ssrd_site(ds, s) for s in SOLAR_FIELDS], ignore_index=True)
all_sites_df.sort_values(['site','time'], inplace=True)
print(all_sites_df.head())
```

```python
# convert accumulated ssrd to hourly energy via positive diffs
def accum_to_hourly_energy(df):
    df = df.copy().sort_values('time')
    hourly = df['ssrd'].diff()
    # handle day/forecast resets (negative/zero steps)
    hourly = hourly.where(hourly > 0)
    return hourly

all_sites_df['ssrd_hourly'] = (
    all_sites_df.groupby('site', group_keys=False).apply(accum_to_hourly_energy)
)

# optional: irradiance (W/m^2) if you need it later
all_sites_df['ssrd_wm2'] = all_sites_df['ssrd_hourly'] / 3600.0

# export the per-site CSVs your analysis uses
outdir = "site_csvs_2021"
os.makedirs(outdir, exist_ok=True)

for site in [s['name'] for s in SOLAR_FIELDS]:
    sub = all_sites_df[all_sites_df['site'] == site].copy()
    sub = sub.dropna(subset=['ssrd_hourly'])
    # keep the minimal schema used downstream: time + ssrd (hourly energy)
    out = sub[['time', 'site', 'ssrd_hourly']].rename(columns={'ssrd_hourly': 'ssrd'})
    outfile = os.path.join(outdir, f"{site}_summer2021.csv")
    out.to_csv(outfile, index=False)
    print(f"[ok] {site}: {len(out)} rows -> {outfile}")
```
